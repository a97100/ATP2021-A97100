Este trabalho, proposto na UC de Algoritmos e Técnicas de Programação foi realizado na linguagem Python. Este consiste num conjunto de funções que fazem a leitura de datasets, do seu processamento e produção de dados estatísticos: distribuicão por modalidade, por curso e por ano; e por fim a visualização destas distribuicões numa janela.

```python
# BdEMD = [EMD]
# EMD = [id, data, nome, idade, genero, morada, modalidade, clube, email, federado, resultado]
# Leitura/carregamento da informação do ficheiro

def convBool(v):
    return v=='true'

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f: 
        novalinha=linha.strip('\n')
        campos=novalinha.split(",")
        emd=[]
        emd.append("emd" + str(campos[1]))
        emd.append(campos[2])
        emd.append(campos[4]+campos[3])
        emd = emd + campos[5:]
        emd[-1]= convBool(emd[-1])
        emd[-2]= convBool(emd[-2])
        bd.append(emd)
        
         
    return bd

BD = []
BD = lerDataset("emd.csv")
print(BD)


# Listagem da informação
def listarDataset(bd): 
    bd.sort(reverse=True, key=lambda emd: emd[1]) #ordenar a data por ordem decrescente
    print("ID      |NOME     |DATA       |RESULTADO")
    print("--------------------------------------------------------")
    for e in bd:
        if e[-1]:
            res='Apto'
        else:
            res='Não Apto'
        print(e[0]  +" | " + e[2] + "  | "  +  e[1] +  " | " +  res)

listarDataset(BD)

# Consulta de um registo
def consultarDataset(bd, id):
    for e in bd:
        if e[0]==id:
            if e[-1]:
                res='Apto'
            else:
                res='Não Apto'
            print(e[0]  +" | " + e[2] + "  | "  +  e[1] +  " | " +  res)

consultarDataset(BD, "emd1")

def modalidades(bd):
    modalidades=[]
    for e in bd:
        if e[6] not in modalidades:
            modalidades.append(e[6])
    modalidades.sort()
    return modalidades

modalidades(BD)

def distribPorModalidade(bd):
    distribuicao={}
    for d in bd:
        if d[6] in distribuicao.keys():
            distribuicao[d[6]]=distribuicao[d[6]]+1
        else:
            distribuicao[d[6]]=1
            
    return distribuicao
distribPorModalidade(BD)

def distribPorClube(bd):
    distribuicao_clube={}
    for d in bd:
        if d[7] in distribuicao_clube.keys():
            distribuicao_clube[d[7]]=distribuicao_clube[d[7]]+1
        else:
            distribuicao_clube[d[7]]=1
            
    return distribuicao_clube
distribPorClube(BD)


def distribPorAno(bd):
    distribuicao={}
    for e in bd:
        ano=e[1][0:4]
        if ano in distribuicao.keys():
            distribuicao[ano]=distribuicao[ano]+1
        else:
            distribuicao[ano]=1
            
    return distribuicao
distribPorAno(BD)
        
def distrib(bd, x):
    distribuicao={}
    for d in bd:
        if d[x] in distribuicao.keys():
            distribuicao[d[x]]=distribuicao[d[x]]+1
        else:
            distribuicao[d[x]]=1
            
    return distribuicao
distrib(BD, 7)

import matplotlib.pyplot as plt

def plotDistribPorModalidade(bd):
    
    distribuicao=distribPorModalidade(bd)

    x = list(range(1,len(distribuicao.keys())+1))
    y = list(distribuicao.values())

    modalidades = list(distribuicao.keys())

    plt.bar(x, y, tick_label = modalidades, width = 0.8, color = ['aquamarine', 'violet', 'pink', 'beige'])

    plt.xlabel('Modalidade')
    plt.ylabel('Nº de alunos')
    plt.title('Distribuição por modalidade')
    plt.xticks(rotation=90)
 

    plt.show()
    
plotDistribPorModalidade(BD)


def plotDistrib(bd, d):
    distribuicao=distrib(bd, d)
 
    x = list(range(1,len(distribuicao.keys())+1))
    y = list(distribuicao.values())

    variavel = list(distribuicao.keys())

    plt.bar(x, y, tick_label = variavel, width = 0.8, color = ['aquamarine', 'violet', 'pink', 'beige'])
    
 plt.ylabel('Nº de alunos')
    plt.title('Distribuição')
    plt.xticks(rotation=90)
 

    plt.show()
    
plotDistrib(BD, 7)



import PySimpleGUI as sg
bd=[]
BD = lerDataset("emd.csv")
bd=BD
dist=[]

menu_list_column = [
    [sg.Button("Distribuição por Clube")],
    [sg.Button("Distribuição por Ano")],
    [sg.Button("Distribuição por Modalidade")],
    [sg.Button("Sair")]
]



data_viewer_column = [
    [sg.Listbox(values=dist,size= (40,20), key="_dist")],
]


layout = [
    [
        sg.Column(menu_list_column),
        sg.VSeperator(),
        sg.Column(data_viewer_column),
    ]
]

window = sg.Window("Distribuições").Layout(layout)


stop = False
while not stop:
    event, values = window.read()
    if event == "Sair" or event == sg.WIN_CLOSED:
        stop = True
    
    elif event == "Distribuição por Ano":
            plotDistrib(bd,2)
            ano=distribPorAno(bd)
            for c in ano.keys():
                dist.append(c +" - " +str(ano[c]) + "\n")
            window.find_element('_dist').update(values=dist)
            dist=[]
            
    elif event == "Distribuição por Clube":  
            plotDistrib(bd,7)
            clube=distribPorClube(bd)
            for c in clube.keys():
                dist.append(c +" - " +str(clube[c]) + "\n")
            window.find_element('_dist').update(values=dist)
            dist=[]
            
    elif event == "Distribuição por Modalidade":  
            plotDistrib(bd,6)
            modalidade=distribPorModalidade(bd)
            for c in modalidade.keys():
                dist.append(c +" - " +str(modalidade[c]) + "\n")
            window.find_element('_dist').update(values=dist)
            dist=[]
        
    else:
        window["-Dados-"].update("Erro: evento desconhecido :: " + inputEvent)

window.close()

```
