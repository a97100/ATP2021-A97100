```python
import matplotlib.pyplot as plt

def menu():
    print(""" 
(1) Criar polinómio.
(2) Calcular o polinómio num ponto.
(3) Fazer uma tabela com valores do polinómio de 1 a n.
(4) Simplificar o polinómio.
(5) Derivar o polinómio.
(6) Desenhar o gráfico de P(x).
(7) Encerrar o programa.""")

    
def criar_termo(coeficiente, expoente):
    return (coeficiente, expoente)


def criar_polinomio():
    return []


def inserir_termo(polinomio, termo):
    return polinomio.append(termo)


def ver_termo(termo):
    coeficiente, expoente = termo
    if expoente == 0:
        return str(coeficiente)
    elif expoente == 1:
        return str(coeficiente)+"x"
    else:
        return str(coeficiente)+"x^"+str(expoente)


def ver_polinomio(polinomio):
    resultado = ""
    for termo in polinomio:
        if resultado == "":
            resultado = ver_termo(termo)
        else:
            resultado = resultado + " " + "+" + " " + ver_termo(termo)
    return print(resultado)


def calcular(polinomio, x):
    resultado=0
    for termo in polinomio:
        coeficiente, expoente = termo
        valor = (x**expoente)*coeficiente
        resultado = resultado + valor
    return resultado


def tabela(polinomio, linhas):
    print("Valor de X :: Valor de p(X)")
    for i in range(linhas):
        return print(str(i) + "::" + str(calcular(polinomio, i)))
        

def derivar(polinomio):
    resultado=[]
    for termo in polinomio:
        coeficiente, expoente=termo
        if expoente>0:
            resultado.append((coeficiente*expoente, expoente-1))
    return resultado


def ordenar(polinomio):
    polinomio.sort(reverse=True, key=chave_ordenada)


def chave_ordenada(termo):
    (coeficiente, expoente)=termo
    return expoente
    

def simplificar(polinomio_ordenado):
    polinomio_novo=[]
    for i in range(1,(len(polinomio_ordenado))):
        coeficiente,expoente=polinomio_ordenado[i]
        coeficiente_0,expoente_0=polinomio_ordenado[i-1]
        if expoente_0==expoente:
            inserir_termo(polinomio_novo,(coeficiente+coeficiente_0,expoente))
        else:
            if expoente_0>expoente:
                if i==1:
                    inserir_termo(polinomio_novo,(coeficiente_0,expoente_0))
            else:
                inserir_termo(polinomio_novo,(coeficiente,expoente))
    return ver_polinomio(polinomio_novo)


def desenhar_polinoio(polinomio):
    
    x=[n for n in range (-100,100, 1)]
    y=[]
    for i in range (-100, 100, 1):    
        y.append(calcular(polinomio, i))
    
    plt.plot(x, y)

    eixoXX=[-100,100]
    eixoXY=[0,0]
    plt.plot(eixoXX, eixoXY, color='black', linewidth = 1)

    eixoYX=[0,0]
    eixoYY=[-100,100]
    plt.plot(eixoYX, eixoYY, color='black', linewidth = 1)

    plt.ylim(-50,50)
    plt.xlim(-50,50)

    plt.xlabel('Abcissas')
    plt.ylabel('Ordenadas')
    plt.title('Um gráfico com estilo!')
    plt.show()

        
variavel="1"
while variavel!="0":
    menu()
    opcao_utilizador=input("Selecione o comando: ")
    
    
    if opcao_utilizador=="7":
        variavel="0"
        print("O programa encerrou. Obrigado e até à próxima.")
    
    
    elif opcao_utilizador=="1":
        tamanho_polinomio=int(input("Tndique o número de  termos do polinomio: "))
        polinomio=criar_polinomio()
        for i in range(tamanho_polinomio):
            coeficiente=int(input("Indique o coeficiente:"))
            expoente=int(input("Indique o expoente: "))
            termo=criar_termo(coeficiente,expoente)
            inserir_termo(polinomio, termo)
        ver_polinomio(polinomio)
    
    
    elif opcao_utilizador=="2":
        x=int(input("Indique o valor de x: "))
        print(calcular(polinomio, x))
    
    
    elif opcao_utilizador=="3":
        linhas=int(input("Indique o número de linhas da tabela: "))
        tabela(polinomio, linhas)
    
    
    elif opcao_utilizador=="4":
        polinomio_ordenado=ordenar(polinomio)
        simplificar(polinomio_ordenado)
    
    
    elif opcao_utilizador=="5":
        polinomio_derivado=derivar(polinomio)
        ver_polinomio(polinomio_derivado)
    
    
    elif opcao_utilizador=="6":
        desenhar_polinoio(polinomio)
```
